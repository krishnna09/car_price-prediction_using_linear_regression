2) Predict price of a BMW X5 that is 10 yr old with mileage 55000

3) Tell me the score (accuracy) of your model. 

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\krishna\Desktop\machine_learning\price_prediction_linear_regression\car_price_prediction - Sheet1.csv")
```

```python
df
```

```python
dummies=pd.get_dummies(df.car_model)
dummies
```

```python
df_dummies=pd.concat([df,dummies],axis='columns')
df_dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['car_model'], axis='columns')
final
```

```python

final = final.drop(['Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop('sell_price', axis='columns')
X
```

```python
Y = final.sell_price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[35000,6,0,0]])
```

```python
model.predict([[55000,10,1,0]])
```

```python
model.score(X,Y)
```
